# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query the ConvertedComp column from the master table
query = "SELECT ConvertedComp FROM master"
df_converted_comp = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Plot the histogram
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.distplot(df_converted_comp['ConvertedComp'].dropna(), bins=30, kde=False)
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.show()

```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query the Age column from the master table
query = "SELECT Age FROM master"
df_age = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Display the first few rows of the dataframe to ensure data is loaded correctly
df_age.head()


import seaborn as sns
import matplotlib.pyplot as plt

# Plot the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_age['Age'].dropna())
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query the Age and WorkWeekHrs columns from the master table
query = "SELECT Age, WorkWeekHrs FROM master"
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Display the first few rows of the dataframe to ensure data is loaded correctly
df.head()

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='WorkWeekHrs', data=df.dropna())
plt.title('Scatter Plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query the WorkWeekHrs, CodeRevHrs, and Age columns from the master table
query = "SELECT WorkWeekHrs, CodeRevHrs, Age FROM master"
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Display the first few rows of the dataframe to ensure data is loaded correctly
df.head()


import seaborn as sns
import matplotlib.pyplot as plt

# Ensure that the Age column doesn't have any negative or zero values for bubble size
df_filtered = df.dropna()
df_filtered = df_filtered[df_filtered['Age'] > 0]

# Plot the bubble plot
plt.figure(figsize=(12, 8))
bubble_plot = sns.scatterplot(
    x='WorkWeekHrs',
    y='CodeRevHrs',
    size='Age',
    data=df_filtered,
    legend=False,  # Disable the legend to avoid clutter
    sizes=(20, 2000),  # Adjust bubble sizes for better visualization
    alpha=0.6  # Make bubbles slightly transparent
)

plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs with Age as Bubble Size')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')

# Add a custom legend for the bubble sizes
for size in [20, 40, 60]:
    plt.scatter([], [], c='b', alpha=0.6, s=size*10,
                label=str(size) + ' Years')

plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='Age')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query the DatabaseDesireNextYear table
query = "SELECT * FROM DatabaseDesireNextYear"
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Count the occurrences of each database
database_counts = df['DatabaseDesireNextYear'].value_counts()

# Get the top 5 databases
top_5_databases = database_counts.head(5)

# Create the pie chart
plt.figure(figsize=(10, 7))
plt.pie(top_5_databases, labels=top_5_databases.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.show()

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query the WorkWeekHrs, CodeRevHrs, and Age columns from the master table
query = "SELECT WorkWeekHrs, CodeRevHrs, Age FROM master"
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Filter the dataframe for the age group 30 to 35
df_filtered = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate the median WorkWeekHrs and CodeRevHrs
median_work_week_hrs = df_filtered['WorkWeekHrs'].median()
median_code_rev_hrs = df_filtered['CodeRevHrs'].median()

# Create a dataframe with the median values
median_df = pd.DataFrame({
    'Metric': ['WorkWeekHrs', 'CodeRevHrs'],
    'Median': [median_work_week_hrs, median_code_rev_hrs]
})

# Plot the stacked bar chart
plt.figure(figsize=(10, 6))
plt.bar(median_df['Metric'], median_df['Median'], color=['blue', 'orange'])
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Metric')
plt.ylabel('Median Hours')
plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query the ConvertedComp and Age columns from the master table
query = "SELECT Age, ConvertedComp FROM master"
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Filter the dataframe for the age group 45 to 60
df_filtered = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate the median ConvertedComp for each age
median_converted_comp = df_filtered.groupby('Age')['ConvertedComp'].median().reset_index()

# Plot the median ConvertedComp for each age
plt.figure(figsize=(10, 6))
plt.plot(median_converted_comp['Age'], median_converted_comp['ConvertedComp'], marker='o')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.show()

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query the MainBranch column from the master table
query = "SELECT MainBranch FROM master"
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Count the occurrences of each category in the MainBranch column
main_branch_counts = df['MainBranch'].value_counts()

# Plot a horizontal bar chart
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='skyblue')
plt.title('Count of Each MainBranch Category')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.show()

```
